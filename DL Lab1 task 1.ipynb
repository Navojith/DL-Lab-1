```python
# 2
import matplotlib.pyplot as plt

plt.hist(np.random.rand(100000), density=True, bins=50, histtype="step", color="blue",
label="random")
plt.hist(np.random.randn(100000), density=True, bins=150, histtype="step", color="red",
label="normal")
plt.hist(np.random.exponential(1,100000), density=True, bins=150, histtype="step", color="green",
label="exp")
plt.axis([-4, 4, 0, 1.5])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# 3
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as matplotlib

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = (X**2 + Y**2 )

plt.figure(1, figsize = (24, 8))
subplot3d = plt.subplot(111, projection='3d')
subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()

```

```python
# 4
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
df.head()

stats_df = df.drop(['Total', 'Stage', 'Legendary','Name',"Type 1", "Type 2"], axis=1)

corr_pearson = stats_df.corr()
corr_spearman = stats_df.corr(method='spearman')

fig, axes = plt.subplots(1, 2, figsize=(14, 8))

sns.heatmap(corr_pearson, ax=axes[0], cmap='coolwarm', annot=True, fmt=".3f")
sns.heatmap(corr_spearman, ax=axes[1], cmap='coolwarm', annot=True, fmt=".3f")

axes[0].set_title('Pearson Correlation')
axes[1].set_title('Spearman Correlation')

plt.show()
```
