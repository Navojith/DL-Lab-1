```python
# 1
import pandas as pd

csv = pd.read_csv("Au_nanoparticle_dataset.csv")
data = csv.loc[:, ['N_total', 'N_bulk', 'N_surface', 'R_avg']]

# 2
data.head(20)
```

```python
# 3
new_data = data.describe().drop(["count","min","max"])
new_data.T
```

```python
# 4
from matplotlib import pyplot as plt

figure, axes = plt.subplots(nrows=1, ncols=4, figsize=(20,5))

N_total = axes[0]
N_bulk = axes[1]
N_surface = axes[2]
R_avg = axes[3]

N_total.hist(data["N_total"], bins=100)
N_total.set_title("N_total")
N_total.set_xlabel("N_total")
N_total.set_ylabel("Frequency")

N_bulk.hist(data["N_bulk"], bins=100)
N_bulk.set_title("N_bulk")
N_bulk.set_xlabel("N_bulk")
N_bulk.set_ylabel("Frequency")

N_surface.hist(data["N_surface"], bins=100)
N_surface.set_title("N_surface")
N_surface.set_xlabel("N_surface")
N_surface.set_ylabel("Frequency")

R_avg.hist(data["R_avg"], bins=100)
R_avg.set_title("R_avg")
R_avg.set_xlabel("R_avg")
R_avg.set_ylabel("Frequency")

plt.show()
```

```python
# 5
import seaborn as sns

sns.pairplot(data, diag_kind="hist")
plt.subplots_adjust(top=0.9)

plt.show()
```

```python
# 6
g = sns.PairGrid(data)
g.map_upper(sns.histplot)
g.map_diag(lambda x, **kws: sns.histplot(x, **kws, kde=True))
g.map_lower(sns.kdeplot, fill=True)

plt.show()
```
